```python
# First, import the relevant modules
import numpy as np
import pandas as pd
from requests.auth import HTTPBasicAuth
import requests
import json
from pandas.io.json import json_normalize
```

```python
def pull_data(url):
    headers = {"Accept": "application/json"}
    auth = HTTPBasicAuth('apikey', API_KEY)
    req = requests.get(url, headers=headers , auth=auth)
    json_obj = None
    if req.status_code == 200:
        json_obj = req.json()
    return json_obj    
```

```python
# Now, call the Quandl API and pull out a small sample of the data (only one day) to get a glimpse
# into the JSON structure that will be returned
url = "https://api.boerse-frankfurt.de/data/price_history?limit=50&offset=0&isin=DE0005313704&mic=XETR&minDate=2020-02-28&maxDate=2020-03-13"
json_obj = pull_data(url)
```

### 1) Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).

```python
url = "https://api.boerse-frankfurt.de/data/price_history?limit=365&offset=0&isin=DE0005313704&mic=XETR&minDate=2017-01-01&maxDate=2017-12-31"
json_obj = pull_data(url)
print(len(json_obj['data']))
json_normalize(json_obj, 'data').sort_values('date').tail(31)
```

### 2) Convert the returned JSON object into a Python dictionary.

```python
# Not sure I follow this question.  Assume it will be a nested dict {date:{prop:value}}
data_by_dates = {}
for day in json_obj['data']:
    date = day.pop('date')
    data_by_dates[date] = day
```

### 3) Calculate what the highest and lowest opening prices were for the stock in this period.

```python
min_open = min([v['open'] for k,v in data_by_dates.items()])
max_open = max([v['open'] for k,v in data_by_dates.items()])
min_open_date = min([k for k,v in data_by_dates.items() if v['open'] == min_open]) #return first date if more then 1
max_open_date = min([k for k,v in data_by_dates.items() if v['open'] == max_open]) #return first date if more then 1
print('Lowest opening price was on {} and opened at {}'.format(min_open_date, min_open))
print('Highest opening price was on {} and opened at {}'.format(max_open_date, max_open))
```

### 4) What was the largest change in any one day (based on High and Low price)?

```python
min_delta = min([v['high'] - v['low'] for k,v in data_by_dates.items()])
max_delta = max([v['high'] - v['low'] for k,v in data_by_dates.items()])
min_delta_date = min([k for k,v in data_by_dates.items() if v['high'] - v['low'] == min_delta])
max_delta_date = min([k for k,v in data_by_dates.items() if v['high'] - v['low'] == max_delta])
print('Lowest intra-day price difference was on {} and opened at {:.4f}'.format(min_delta_date, min_delta))
print('Highest intra-day price difference  was on {} and opened at {:.4f}'.format(max_delta_date, max_delta))
```

### 5) What was the largest change between any two days (based on Closing Price)?

```python
previous_close = -1
max_close_change = 0
max_clost_change_date = '1/1/2017'
for day in sorted(data_by_dates.keys()):
    current_close = data_by_dates[day]['close']
    if previous_close == -1:
        max_clost_change_date = day
        previous_close = current_close
    else:
        if np.abs(current_close - previous_close) > max_close_change:
            max_close_change = current_close - previous_close
            max_clost_change_date = day
        
        previous_close = current_close
print('Biggest change in closing prices (up or down) as on {} and with a change of {:.4f}'
      .format(max_clost_change_date, max_close_change))      
```

### 6) What was the average daily trading volume during this year?

```python
mean_volumn = np.mean([v['turnoverPieces'] for k,v in data_by_dates.items()])
print('The mean trading volume for 2017 was {:,.0f}'.format(mean_volumn))
```

### 7) (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
median_volumn = np.median([v['turnoverPieces'] for k,v in data_by_dates.items()])
print('The median trading volume for 2017 was {:,.0f}'.format(median_volumn))
```
