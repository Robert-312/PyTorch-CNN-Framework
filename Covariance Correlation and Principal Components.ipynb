```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mu, sigma = 1, 0.1 # mean and standard deviation
x = np.random.normal(mu, sigma, 1000)
mu = .5
y_add = np.random.normal(mu, sigma, 1000)
df = pd.DataFrame(x, columns=['x'])
df['y'] = x + y_add
df.hist(bins=100)
plt.show()
```

```python
df_cov = pd.DataFrame(df.cov())
print(df_cov)

df_corr = pd.DataFrame(df.corr())
print(df_corr)
```

```python
df.plot(kind='scatter', x='x', y='y')
plt.show()
```

```python
df_transpose = pd.DataFrame(df.transpose())
pc = df_corr.dot(df_transpose)
pc = pc.transpose()
pc.columns = ['xt','yt']
print(pc)
pc.plot(kind='scatter', x='xt', y='yt')
plt.show()
```
